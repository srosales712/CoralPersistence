```python
CF_bottom=pd.DataFrame.from_dict(obj['CF']["Bottom"])
CF_bottom['reef'] = 'cf'
CF_bottom['place'] = 'bottom'
CF_bottom.columns = CF_top.columns.str.replace(' ', '') 
CF_bottom
```

```python
ND_bottom=pd.DataFrame.from_dict(obj['ND']["Bottom"])
ND_bottom['reef'] = 'ndr'
ND_bottom['place'] = 'bottom'
ND_bottom.columns = ND_bottom.columns.str.replace(' ', '') 

ND_Top=pd.DataFrame.from_dict(obj['ND']["Top"])
ND_Top['reef'] = 'ndr'
ND_Top['place'] = 'top'
ND_Top.columns = ND_Top.columns.str.replace(' ', '') 

ND_sand=pd.DataFrame.from_dict(obj['ND']["Sand"])
ND_sand['reef'] = 'ndr'
ND_sand['place'] = 'sand'
ND_sand.columns = ND_sand.columns.str.replace(' ', '') 
```
